```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
from pathlib import Path
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func, inspect, text

from flask import Flask, jsonify

```

```python
path = Path("spotifiy_data/MostStreamedSpotifySongs2024.csv")
df = pd.read_csv(path, encoding="ISO-8859-1")
df.head(100)
```

```python
df = df.map(lambda x: x.replace(',', '') if isinstance(x, str) else x)
```

```python
# df.to_csv("spotify_cleaned.csv")
```

```python
engine = create_engine("sqlite:///spotify.sqlite")
```

```python
engine.connect()
```

```python
inspector = inspect(engine)
```

```python
df.to_sql("spotify_data", engine, index=True, if_exists='replace')
```

```python
session = Session(engine)
```

```python
Base = automap_base()
Base.prepare(autoload_with=engine)
```

```python
inspector.get_table_names()
```

```python
Base.classes.keys()
```

```python
with engine.connect() as con:
    con.execute(text('ALTER TABLE spotify_data ADD PRIMARY KEY (Track);'))
```
