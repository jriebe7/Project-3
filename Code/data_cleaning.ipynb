```python
import pandas as pd
```

```python
df = pd.read_csv("Most Streamed Spotify Songs 2024.csv", encoding='ISO-8859-1')
df.head(10)
```

```python
print(df.dtypes)
```

```python
print(df.columns)
```

```python
# Remove commas from the column
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].astype(str).str.replace(',', '')

# Convert the column to float64 data type
df['Spotify Playlist Count'] = df['Spotify Playlist Count'].astype(float)

# Replace NaN values with 0 in the following columns
columns_to_fill = ['Spotify Playlist Count', 'Apple Music Playlist Count', 'Deezer Playlist Count', 'Amazon Playlist Count']

df[columns_to_fill] = df[columns_to_fill].fillna(0)
```
