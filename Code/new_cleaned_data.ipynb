```python
import pandas as pd
```

```python
df = pd.read_csv("../Cleaned_Datasets/cleaned_data_visual_two.csv", encoding='ISO-8859-1')
df.head(10)
```

```python
# Remove commas from the columns
columns_to_clean = ['Spotify Streams', 'YouTube Views', 'TikTok Views', 'Pandora Streams']
df[columns_to_clean] = df[columns_to_clean].astype(str).apply(lambda x: x.str.replace(',', ''))

# Convert the columns to float64 data type
# df[columns_to_clean] = df[columns_to_clean].astype(int)

# Replace 'nan' values with None (NULL) in the specified columns
pd.set_option('display.float_format', lambda x: '%.0f' % x)
columns_to_fill = ['Spotify Streams', 'YouTube Views', 'TikTok Views', 'Pandora Streams']
df[columns_to_fill] = df[columns_to_fill].apply(lambda x: pd.to_numeric(x, errors='coerce'))

# Fill NaN values with 0 after conversion
df[columns_to_fill] = df[columns_to_fill].fillna(0)
df[columns_to_fill] = df[columns_to_fill].astype(int).astype(str)
```

```python
cleaned_dataset_df = df
```
